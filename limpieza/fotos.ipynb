```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Lee el archivo CSV
csv_path = r"C:\Users\Rembert\Documents\GitHub\Proyecto-Final\csv_filtrados\cleaned_player.csv"
df = pd.read_csv(csv_path)

# Extrae la lista de nombres completos
player_names = df['Full Name'].tolist()

def get_player_image_url(player_name):
    # Convierte el nombre a formato adecuado para URL
    split_name = player_name.split()
    if len(split_name) < 2:
        return None

    last_name = split_name[1].lower()
    first_initial = split_name[0][0].lower()
    
    # Intenta varios sufijos para encontrar la URL correcta
    for i in range(1, 10):
        search_name = f"{last_name}{first_initial}{i:02d}"
        search_url = f"https://www.basketball-reference.com/players/{first_initial}/{search_name}.html"
        
        print(f"Buscando en URL: {search_url}")

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
        }
        response = requests.get(search_url, headers=headers)
        if response.status_code != 200:
            print(f"No se pudo acceder a la página de {player_name}. Código de estado: {response.status_code}. Intentando siguiente sufijo.")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Encuentra la imagen del jugador
        img_tag = soup.find('img', {'class': 'media-item'})
        if img_tag:
            img_url = img_tag['src']
            return img_url

    return None

# Lista para almacenar los resultados
results = []

# Procesa la lista de nombres de jugadores
for player in player_names:
    print(f"Buscando imagen para {player}...")
    img_url = get_player_image_url(player)
    results.append({'Player Name': player, 'Image URL': img_url})
    if img_url:
        print(f"Imagen encontrada para {player}: {img_url}")
    else:
        print(f"No se encontró imagen para {player}.")
    
    time.sleep(5)  # Espera 5 segundos entre solicitudes

# Convierte la lista de resultados en un DataFrame de Pandas
results_df = pd.DataFrame(results)

# Guarda el DataFrame en un archivo CSV
output_csv_path = r"C:\Users\Rembert\Documents\GitHub\Proyecto-Final\csv_filtrados\player_images.csv"
results_df.to_csv(output_csv_path, index=False)

print(f"Archivo CSV generado: {output_csv_path}")
```
