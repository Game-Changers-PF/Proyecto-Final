```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import commonplayerinfo
import pandas as pd
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.endpoints import leaguegamefinder, boxscoretraditionalv2
from nba_api.stats.endpoints import shotchartdetail
import time
```

```python
from datetime import datetime
```

```python
df_calendario = pd.read_csv("/Users/usuario1/Documents/GitHub/Proyecto-Final/csv_originales/calendario_partidos_2015_2022.csv")

df_calendario.head()
```

```python
# Convertir la columna GAME_DATE a tipo datetime si no estÃ¡ en ese formato
df_calendario['GAME_DATE'] = pd.to_datetime(df_calendario['GAME_DATE'])

# Filtrar por fechas especÃ­ficas (por ejemplo, solo fechas despuÃ©s de '2023-01-01')
fecha_limite = datetime.strptime('2015-01-01', '%Y-%m-%d')
df_filtrado = df_calendario[df_calendario['GAME_DATE'] > fecha_limite]

df_filtrado.head()
```

```python
# columnas que no aportan informaciÃ³n relevante fueron eliminadas
columnas_no_relevantes = ['MATCHUP', 'MIN', 'WL', 'PLUS_MINUS']
df_procesado = df_filtrado.drop(columns=columnas_no_relevantes)

```

```python
# AquÃ­ es donde guardaremos el DataFrame procesado en un nuevo archivo CSV
ruta_nuevo_csv = '/Users/usuario1/Documents/GitHub/Proyecto-Final/csv_filtrados/cleaned_calendario_partidos_2015_2022.csv'

# Guardar el DataFrame procesado en un nuevo archivo CSV
df_procesado.to_csv(ruta_nuevo_csv, index=False)

print(f"Archivo guardado correctamente en {ruta_nuevo_csv}")
```
