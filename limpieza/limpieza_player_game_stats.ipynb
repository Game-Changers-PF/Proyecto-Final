```python
import numpy as np
import pandas as pd
```

```python
# leemos el archivo csv "player_game_stats_2015_2022_detailed.csv" con los datos necesarios 
player_game_stats = pd.read_csv(r"C:\Users\Lenovo\Desktop\proyectoFina\Proyecto-Final\csv_originales\player_game_stats_2015_2022_detailed.csv", sep=",")
```

```python
# Borrar columnas que no necesitariamos, en este caso NICKNAME , PLAYER_NAME
player_game_stats = player_game_stats.drop(columns=['NICKNAME' , 'PLAYER_NAME'])
```

```python
print(player_game_stats)
```

```python
#De la columna "season" modificamos el "-NN", indicando que solo queremos ver los años que aparecen en esta columna
player_game_stats['SEASON'] = player_game_stats['SEASON'].str.replace(r'-\d{2}$', '', regex=True)
```

```python
print(player_game_stats)
```

```python
#queremos ver los datos duplicados dentro de la tabla
duplicated_rows = player_game_stats.duplicated()
```

```python
# visualizamos los datos duplicados
duplicated_data = player_game_stats.loc[duplicated_rows]
print(duplicated_data)
```

```python
# Contar los valores nulos en las columnas específicas
columns_to_check = ['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT',
                    'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB',
                    'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS', 'PLUS_MINUS']
null_counts = player_game_stats[columns_to_check].isnull().sum()

print("Valores nulos por columna:")
print(null_counts)
```

```python
# Definir la lista de columnas a revisar
columns_to_check = ['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT',
                    'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB',
                    'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS', 'PLUS_MINUS']

# Contar los valores nulos en las columnas específicas
null_counts = player_game_stats[columns_to_check].isnull().sum()

print("Valores nulos por columna:")
print(null_counts)

# Llenar los valores nulos con 0 en las columnas específicas
player_game_stats[columns_to_check] = player_game_stats[columns_to_check].fillna(0)

# Contar los valores nulos nuevamente para verificar
null_counts_after_fill = player_game_stats[columns_to_check].isnull().sum()

print("Valores nulos por columna después de llenar con 0:")
print(null_counts_after_fill)
```

```python
#Se observan que la columna posee datos decimales, se modifican los datos decimales pasando a convertirlos a enteros
def convertir_a_formato_deseado(valor):
    try:
        minutos_decimales, segundos = valor.split(':')
        minutos_entero = int(float(minutos_decimales))  # Convertir minutos a entero
        segundos_entero = int(segundos)     # Convertir segundos a entero
        return f"{minutos_entero}:{segundos_entero:02}"  # Formato MM:SS
    except Exception as e:
        print(f"Error al convertir {valor}: {e}")
        return None

# Aplicar la función a la columna MIN
player_game_stats['MIN'] = player_game_stats['MIN'].apply(convertir_a_formato_deseado)

print(player_game_stats)
```

```python
# Rellenar los valores nulos en la columna 'MIN' con '0:00'
player_game_stats['MIN'] = player_game_stats['MIN'].fillna('0:00')

# Verificar si todavía hay valores nulos en la columna 'MIN'
null_counts_min = player_game_stats['MIN'].isnull().sum()

print("Valores nulos en la columna 'MIN' después de llenar con '0:00':")
print(null_counts_min)
```

```python
#Consultar formato de todas las columnas de la tabla
print(player_game_stats.dtypes)
```

```python
#Al obtener datos donde se observan inconsistencias, ya que algunos datos poseen un formato M:SS y se desea que todos los datos esten en formato MM:SS
player_game_stats['MIN'] = player_game_stats['MIN'].apply(lambda x: ':'.join([f"{int(part):02d}" for part in x.split(':')]))
```

```python

null_counts_df_cleaned = player_game_stats.isnull().sum()

print('valores nulos por columnas:')
print(null_counts_df_cleaned)
```

Un dato que aparece al buscar valores nulos, podemos observar que hay varios jugadores que no parten de una posicion inicial, es decir que no poseen una posicion fija, sino que son versatiles y tambien esto puede hacer que tengan estadisticas mas variadas con respecto a su juego.

```python
# Rellenar los valores nulos en la columna 'COMMENT' con 'no comments'
player_game_stats['COMMENT'] = player_game_stats['COMMENT'].fillna('no comments')

# Verificar si todavía hay valores nulos en la columna 'COMMENT'
null_counts_comment = player_game_stats['COMMENT'].isnull().sum()

print("Valores nulos en la columna 'COMMENT' después de llenar con 'no comments':")
print(null_counts_comment)
```

```python
# Rellenar los valores nulos en la columna 'START_POSITION' con 'Bench'
player_game_stats['START_POSITION'] = player_game_stats['START_POSITION'].fillna('NA')

# Verificar si todavía hay valores nulos en la columna 'START_POSITION'
null_counts_start_position = player_game_stats['START_POSITION'].isnull().sum()

print("Valores nulos en la columna 'START_POSITION' después de llenar con 'NA':")
print(null_counts_start_position)
```
